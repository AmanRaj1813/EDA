DATA **CLEANING**

```python
print(f'id : {taxi_df.id.isnull().sum()}')
print(f'vendor_id : {taxi_df.vendor_id.isnull().sum()}')
print(f'pickup_datetime : {taxi_df.pickup_datetime.isnull().sum()}')
print(f'dropoff_datetime : {taxi_df.dropoff_datetime.isnull().sum()}')
print(f'passenger_count : {taxi_df.passenger_count.isnull().sum()}')
print(f'pickup_longitude : {taxi_df.pickup_longitude.isnull().sum()}')
print(f'pickup_latitude : {taxi_df.pickup_latitude.isnull().sum()}')
print(f'dropoff_longitude : {taxi_df.dropoff_longitude.isnull().sum()}')
print(f'dropoff_latitude : {taxi_df.dropoff_latitude.isnull().sum()}')
print(f'store_and_fwd_flag : {taxi_df.store_and_fwd_flag.isnull().sum()}')
print(f'trip_duration : {taxi_df.trip_duration.isnull().sum()}')
```

```python
taxi_df_specific_column = taxi_df.copy()
taxi_df_specific_column['dropoff_datetime'] = taxi_df_specific_column['dropoff_datetime'].fillna(0)
taxi_df_specific_column['passenger_count'] = taxi_df_specific_column['passenger_count'].fillna(0)
taxi_df_specific_column['pickup_longitude'] = taxi_df_specific_column['pickup_longitude'].fillna(0)

taxi_df_specific_column['pickup_latitude'] = taxi_df_specific_column['pickup_latitude'].fillna(0)
taxi_df_specific_column['dropoff_longitude'] = taxi_df_specific_column['dropoff_longitude'].fillna(0)
taxi_df_specific_column['dropoff_latitude'] = taxi_df_specific_column['dropoff_latitude'].fillna(0)
taxi_df_specific_column['store_and_fwd_flag'] = taxi_df_specific_column['store_and_fwd_flag'].fillna(0)
taxi_df_specific_column['trip_duration'] = taxi_df_specific_column['trip_duration'].fillna(0)
```

```python
print(taxi_df_specific_column.isnull().sum())
```

```python
#Removing zeros from the columns requied for fi=urther analysis
taxi_df = taxi_df[taxi_df['trip_duration']>0]
taxi_df = taxi_df[taxi_df['passenger_count']>0]
taxi_df = taxi_df[(taxi_df['pickup_latitude']!=0)&(taxi_df['pickup_longitude']!=0) &(taxi_df['dropoff_latitude']!=0)&(taxi_df['dropoff_longitude']!=0)]
taxi_df.shape
```

**Data Transformation and Feature Engineering**

```python
## converting pickup_datetime and dropoff_time from object data type to DateTime data type for extraction of date time month year
taxi_df['pickup_datetime'] = pd.to_datetime(taxi_df['pickup_datetime'], errors='coerce')
taxi_df['dropoff_datetime'] = pd.to_datetime(taxi_df['dropoff_datetime'], errors='coerce')

print("DataFrame after converting 'pickup_datetime,dropoff_time' to datetime dtype:")
print(taxi_df.dtypes['pickup_datetime'])
print(taxi_df.dtypes['dropoff_datetime'])
```

```python
## For pickup_datetime
##year
taxi_df['p_year'] = taxi_df['pickup_datetime'].dt.year
##month
taxi_df['p_month'] = taxi_df['pickup_datetime'].dt.month
taxi_df['p_monthName'] = taxi_df['pickup_datetime'].dt.month_name()
##date
taxi_df['p_dayName'] = taxi_df['pickup_datetime'].dt.day_name()
## Time
taxi_df['p_time'] = taxi_df['pickup_datetime'].dt.time
##Hour
taxi_df['p_hour'] = taxi_df['pickup_datetime'].dt.hour
##Mins
taxi_df['p_mins'] = taxi_df['pickup_datetime'].dt.minute
```

```python
## For dropoff_datetime
##year
taxi_df['d_year'] = taxi_df['dropoff_datetime'].dt.year
##month
taxi_df['d_month'] = taxi_df['dropoff_datetime'].dt.month
taxi_df['d_monthName'] = taxi_df['dropoff_datetime'].dt.month_name()
##date
taxi_df['d_dayName'] = taxi_df['dropoff_datetime'].dt.day_name()
## Time
taxi_df['d_time'] = taxi_df['dropoff_datetime'].dt.time
##Hour
taxi_df['d_hour'] = taxi_df['dropoff_datetime'].dt.hour
##Mins
taxi_df['d_mins'] = taxi_df['dropoff_datetime'].dt.minute
```

**Correlatiom**

```python
from scipy.stats import pearsonr
corr_1, pValue_1 = pearsonr(taxi_df["pickup_latitude"], taxi_df['dropoff_latitude'])
corr_2, pValue_2 = pearsonr(taxi_df["p_hour"], taxi_df['d_hour'])
corr_3, pValue_3 = pearsonr(taxi_df["passenger_count"], taxi_df['trip_duration'])


print('Pearsons Correlation:')
print('between pickup_lattitude and dropoff_lattitude is %.3f' % corr_1)
print('between Pickup hour and Dropoff hour is %.3f' % corr_2)
print('between Passenger Count and trip duration is %.3f' % corr_3)
```

**Inference :**


*   The pickup hour and the dropoff hour are correlated better so we can say that the model predicting the pickup and dropoff time is working in the best predictable way.
*   The correlation between passenger count and trip duration is very bad with this we can interpret that for long duration journey, people of nyc don't prefer Taxi services.




## **Time-series insights**

**HeatMap to understand the Taxi's damand in days and months**

```python
trip_counts_month_day = taxi_df.groupby(['p_monthName', 'p_dayName']).size().unstack(fill_value=0)
#maping month and day for readability
month_order = [
    'January', 'February', 'March', 'April', 'May', 'June']
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

trip_counts_month_day = trip_counts_month_day.reindex(month_order, axis=0)
trip_counts_month_day = trip_counts_month_day.reindex(day_order, axis=1)
plt.figure(figsize=(15, 9)) # Adjust figure size for better readability
sns.heatmap(
    trip_counts_month_day,
    annot=True,     # Show the counts on the heatmap cells
    fmt='f' ,
    cmap='YlGnBu',
    linewidths=.5,
    cbar_kws={'label': 'Number of Taxi Trips'}
)
plt.title('Taxi Trip Volume: Month vs. Day of Week', fontsize=16)
plt.xlabel('Day of Week', fontsize=14)
plt.ylabel('Month', fontsize=14)
plt.xticks(rotation=45, ha='right') # Rotate day names for better fit if needed
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

```

**Inference** : This heatmap shows darker the shade more the trips which is observed on **fridays** and **saturdays** of every month and the leading month is **April** with maximum trips followed by **may** .

```python
# Line plot for pickup and dropoff date
plt.figure(figsize=(14, 7))
sns.lineplot(data=taxi_df, x=taxi_df.pickup_datetime, y='dropoff_datetime', color='blue')
plt.title('pickup datetime over dropoff datetime')
plt.xlabel('pickup Date')
plt.ylabel('Dropoff Date')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Inference** : Mostly the trip completes on the same day itself as the plot varies linearly in strainght line.

```python
# Line plot for pickup and dropoff time find max/min trip duration.
plt.figure(figsize=(14, 7))
sns.lineplot(data=taxi_df, x=taxi_df.p_mins, y='d_mins', color='blue')
plt.title('pickup time over dropoff time')
plt.xlabel('Pickup time')
plt.ylabel('Dropoff time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Inference** : The peak of this plot show that the trip duration is maximum over that time where as lower left hand and right hand side interprets that trip duration was comparatively less.

## **Business Insights**

**FOR PEAK HOURS IN A DAY**

```python
peak_hours_counts = taxi_df['p_hour'].value_counts().sort_index()
print("Trip counts per hour of the day:")
print(peak_hours_counts.max())
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x=peak_hours_counts.index, y=peak_hours_counts.values, palette='plasma' )
plt.title('Number of Trips by Hour of Day (Peak Hours)')
plt.xlabel('Hour of Day (0-23)')
plt.ylabel('Number of Trips')
plt.xticks(np.arange(0, 24, 1)) # Ensure all 24 hours are shown on x-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

**Inference** :

*   18th and 19th hrs of the day is the peak pickup time for taxi's so they can
make good business in those time period and least in 4th ans 5th hrs of the day so these distribution can be considered.
*   This also interprets that people prefer to travel more during evening and nights.






**FOR PEAK DAY IN A WEEK**

```python
peak_dayname_counts = taxi_df['p_dayName'].value_counts().sort_index()
print("Trip counts per day of the week:")
print(peak_dayname_counts)
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
peak_dayname_counts = peak_dayname_counts.reindex(day_order, axis=1)
```
